```python
# Import all world cup files and concatenate them in one dataframe name as all_wc_data
import os
import pandas as pd
file_path = "D:\Python Pandas Data\Football Matches"
files = os.listdir(file_path)

all_wc_data = pd.DataFrame()
for file in files:
    current_data = pd.read_csv(file_path+'/'+file)
    current_data['Year'] = file
    current_data['Year']= current_data['Year'].str[7:11]
    all_wc_data = pd.concat([all_wc_data,current_data])

all_wc_data #call the dataframe
```

```python
#import other file that contains summary of the Fifa world cup from 1930 to 2022
fifa_wc_sum = pd.read_csv(r"D:\Python Pandas Data\FIFA - World Cup Summary.csv")
fifa_wc_sum
```

```python
#Adding new column that calculates total goals happened in a worlcup matches according to teams and worldcup year
all_wc_data['Total Goals'] = all_wc_data['Goals For'] + all_wc_data['Goals Against']
```

```python
#Summary of all columns grouby Team
all_wc_data.groupby(['Team'])[['Games Played','Win','Loss','Draw','Total Goals','Goals For','Goals Against']].sum().sort_values(by='Games Played', ascending=False)
```

```python
#Summary of total goals (including Goals for & against) by each team
all_wc_data.groupby(['Team'])[['Total Goals','Goals For','Goals Against']].sum().sort_values(by='Total Goals', ascending=False)
```

```python
#Summary of total matches played & goals occured during each Fifa world cup

all_wc_data.groupby(['Year'])[['Games Played','Total Goals']].sum()
```

```python
#Bar chart of total matches played & goals occured during each Fifa world cup

import matplotlib.pyplot as plt
all_wc_data.groupby(['Year'])[['Games Played','Total Goals']].sum().plot.bar(figsize=[12,8])
plt.ylabel('Total Goals')
plt.xlabel('Year')
plt.title('Total Games Played & Goals from 1930 to 2022')
```

```python
#Summary of total matches played & number of win & loss (with win & loss percentage) by all teams in all Fifa world cup matches from 1930 to 2022

tot_game_played = all_wc_data.groupby(['Team'])[['Games Played','Win','Loss','Draw']].sum()
tot_game_played['Win%'] = ((tot_game_played['Win']/tot_game_played['Games Played']).apply('{:.0%}'.format))
tot_game_played['Loss%'] = ((tot_game_played['Loss']/tot_game_played['Games Played']).apply('{:.0%}'.format))
tot_game_played['Draw%'] = ((tot_game_played['Draw']/tot_game_played['Games Played']).apply('{:.0%}'.format))

tot_game_played.sort_values(by='Win%', ascending=False)
```

```python
#Bar chart of total matches played & number of win & loss by top 20 teams in all Fifa world cup matches from 1930 to 2022

all_wc_data.groupby(['Team'])[['Games Played','Win','Loss']].sum().sort_values(by='Games Played', ascending=False).head(20).plot.bar(figsize=[12,8])
plt.xlabel('Top 20 Countries Played Most Number of Matches')
plt.title('Top 20 Countries Played Most Matches Since 1930 with Win & Loss Number of Matches')
```

```python
#Countries with most matches winner by percentage
tot_game_played.filter(items = ['Team','Win%','Loss%','Draw%'],axis=1).sort_values(by='Win%', ascending=False)
```

```python
champ = fifa_wc_sum.groupby(['CHAMPION'])['CHAMPION'].count()
runn = fifa_wc_sum.groupby(['RUNNER UP'])['RUNNER UP'].count()
third = fifa_wc_sum.groupby(['THIRD PLACE'])['THIRD PLACE'].count()
```

```python
#Summary of each team with how many time they were Chapion, Runner up & at Third place
pd.concat([champ,runn,third],axis=1).fillna('')
```

```python
#Bar chart of each team with how many time they were Chapion, Runner up & at Third place

pd.concat([champ,runn,third],axis=1).sort_values(by='CHAMPION',ascending=False).plot.bar(figsize=[12,8])
plt.title('Champions, Runner up & Third Place Summary')
plt.xlabel('Countries')
plt.ylabel('No. of Champion, Runner up & Third Place')
```
